Cleaning and transforming the gathered information

```python
# Getting rid of trial races as they are irrelevent
def allign_columns(race_info): 
    if isinstance(race_info, int):
        return race_info
    
    race_info_parts = str(race_info).split(' ')
    for part in race_info_parts:
        if 'TRIAL' in part:
            return 0
    
    return ' '.join(race_info_parts)

df_word['race1'] = df_word['race1'].apply(lambda x: allign_columns(x))
df_word['race2'] = df_word['race2'].apply(lambda x: allign_columns(x))
df_word['race3'] = df_word['race3'].apply(lambda x: allign_columns(x))
df_word['race4'] = df_word['race4'].apply(lambda x: allign_columns(x))
df_word['race5'] = df_word['race5'].apply(lambda x: allign_columns(x))
df_word['race6'] = df_word['race6'].apply(lambda x: allign_columns(x))
df_word['race7'] = df_word['race7'].apply(lambda x: allign_columns(x))

# swapping from the last races as only the information from recent 3 races will be collected
columns_to_swap = ['race1', 'race2', 'race3', 'race4', 'race5', 'race6', 'race7']
def swap_zeros(df):
    swapped = False
    for i in range(len(columns_to_swap) - 1):
        current_col = columns_to_swap[i]
        next_col = columns_to_swap[i + 1]
        zero_indices = df[df[current_col] == 0].index
        for index in zero_indices:
            if df.loc[index, next_col] != 0:
                df.loc[index, current_col], df.loc[index, next_col] = df.loc[index, next_col], df.loc[index, current_col]
                swapped = True
    return df, swapped

# Apply the swapping logic iteratively until columns 1, 2, and 3 have no zero values
has_zeros = True
while has_zeros:
    df_word, has_zeros = swap_zeros(df_word)

#extracting the dates of the races and transforming them into weeks
def extract_date(race_info):
    if race_info == 0:
        return 0
    else:
        race_info_parts = race_info.split(' ')
        for part in race_info_parts:
            if '/' in part:
                date_part = part.split('\n')[1]
                return date_part


df_word['date1'] = df_word['race1'].apply(lambda x: extract_date(x))
df_word['date2'] = df_word['race2'].apply(lambda x: extract_date(x))
df_word['date3'] = df_word['race3'].apply(lambda x: extract_date(x))

date_format = '%d/%m/%Y'
df_word['Date1'] = pd.to_datetime(df_word['date1'], format=date_format, errors='coerce')
df_word['Date2'] = pd.to_datetime(df_word['date2'], format=date_format, errors='coerce')
df_word['Date3'] = pd.to_datetime(df_word['date3'], format=date_format, errors='coerce')

today = datetime.today()
weeks_passed = lambda date: (today - date).days // 7 if not pd.isnull(date) else 0

df_word['weeks_passed1'] = df_word['Date1'].apply(weeks_passed)
df_word['weeks_passed2'] = df_word['Date2'].apply(weeks_passed)
df_word['weeks_passed3'] = df_word['Date3'].apply(weeks_passed)

columns_to_fill = ['position_r1', 'position_r2', 'position_r3',
       'position_r4', 'position_r5', 'position_r6', 'position_r7']

df_word[columns_to_fill]=df_word[columns_to_fill].fillna(0)
df_word[columns_to_fill]=df_word[columns_to_fill].replace('', 0)

columns_to_swap_length = ['position_r1', 'position_r2', 'position_r3',
       'position_r4', 'position_r5', 'position_r6', 'position_r7']
def swap_zeros_length(df):
    swapped = False
    for i in range(len(columns_to_swap_length) - 1):
        current_col = columns_to_swap_length[i]
        next_col = columns_to_swap_length[i + 1]
        zero_indices = df[df[current_col] == 0].index
        for index in zero_indices:
            if df.loc[index, next_col] != 0:
                df.loc[index, current_col], df.loc[index, next_col] = df.loc[index, next_col], df.loc[index, current_col]
                swapped = True
    return df, swapped

# Apply the swapping logic iteratively until columns 1, 2, and 3 have no zero values
has_zeros_length = True
while has_zeros_length:
    df_word, has_zeros_length = swap_zeros_length(df_word)

#Cleaning the columns so only the position and length behind the winner remains

def cleaning_columns(column_1):
    hh1 = []
    for i in df_word[column_1]:
        if i == 0:
            hh1.append(0)
        else:    
            i_split=i.split('\n')[0]
            hh=i_split.split(' ')[0]
            hh1.append(hh)
    return hh1


def cleaning_columns2(column_2):
    hh2 = []
    for ix in df_word[column_2]:
        if ix == 0:
            hh2.append(0)
        else:
            try:
            
                i_split2=ix.split('\n')[2]
                hh12 = i_split2.replace('L', '')
                hh2.append(hh12)

            except:
                hh2.append(0)


    return hh2

df_word['position_r1[1]'] = cleaning_columns('position_r1')
df_word['position_r2[1]'] = cleaning_columns('position_r2')
df_word['position_r3[1]'] = cleaning_columns('position_r3')

df_word['position_r1[2]'] = cleaning_columns2('position_r1')
df_word['position_r2[2]'] = cleaning_columns2('position_r2')
df_word['position_r3[2]'] = cleaning_columns2('position_r3')
```
