```python
#pip install the required libraries if needed
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time

```

```python
website_racenet = 'https://www.racenet.com.au/form-guide/horse-racing/caulfield-20230722/the-big-screen-company-bletchingly-stakes-race-8' #insert webpage links
website_ladbrokes = 'https://www.ladbrokes.com.au/racing/caulfield/1abf64f6-32a7-4901-9b41-0a67a628fcb9'
path = r"***" # insert file directory for chromeriver according to your chrome version
```

```python
driver = webdriver.Chrome(path)
driver.get(f'{website_racenet}'+ '/full-form')

field_condition = []
weather_cond = driver.find_elements_by_class_name('meeting-selector__weather')
for x in weather_cond:
    field_condition.append(x.find_element_by_tag_name('span').text)

info = []
main_search = driver.find_elements_by_css_selector("[class='selection-details']")
name_2 = []
elements = []
for main in main_search:
    name_1 = main.find_elements_by_class_name('selection-details__details')
    
    for g in name_1:
        name_2.append(g.find_element_by_tag_name('strong').text)
    for x in name_1:
        elements.append(g.find_element_by_tag_name('span').text)

    
stats_search = driver.find_elements_by_class_name('horseracing-selection-stats')
for car in stats_search:
    career = car.find_elements_by_id('filter-career')
    for carr in career:
        career_number = carr.find_element_by_class_name('form-grid-box__details').text

    last_10 = car.find_elements_by_id('filter-last10')
    for i in last_10:
        last_10_11 = i.find_element_by_class_name('form-grid-box__details').text
    #win_rate = ''
    winrate = car.find_elements_by_id('filter-wins')
    for i in winrate:
        win_rate = i.find_element_by_class_name('form-grid-box__details').text 
    #place = ''
    place_draft = car.find_elements_by_id('filter-places')
    for i in place_draft:
        place = i.find_element_by_class_name('form-grid-box__details').text 
        
    #track = ''
    track_draft = car.find_elements_by_id('filter-track')
    for i in track_draft:
        track = i.find_element_by_class_name('form-grid-box__details').text 
    #distance = ''
    distance_draft = car.find_elements_by_id('filter-distance')
    for i in distance_draft :
        distance = i.find_element_by_class_name('form-grid-box__details').text    
    #track_distance =  ''
    track_distance_draft = car.find_elements_by_id('filter-trackDistance')
    for i in track_distance_draft:
        track_distance = i.find_element_by_class_name('form-grid-box__details').text    
       
    #good = []
    good_draft = car.find_elements_by_id('filter-good')
    for i in good_draft:
        good =i.find_element_by_class_name('form-grid-box__details').text
    #soft = ''
    soft_draft = car.find_elements_by_id('filter-soft')
    for i in soft_draft:
        soft = i.find_element_by_class_name('form-grid-box__details').text
    #heavy = ''
    heavy_draft = car.find_elements_by_id('filter-heavy')
    for i in heavy_draft:
        heavy = i.find_element_by_class_name('form-grid-box__details').text
    #wet = ''
    wet_draft = car.find_elements_by_id('filter-wet')
    for i in wet_draft:
        wet = i.find_element_by_class_name('form-grid-box__details').text

    #firstup = ''
    first_draft = car.find_elements_by_id('filter-firstUp')
    for i in first_draft:
        firstup = i.find_element_by_class_name('form-grid-box__details').text
    

    second_draft = car.find_elements_by_id('filter-secondUp')
    for i in second_draft:
        secondup = i.find_element_by_class_name('form-grid-box__details').text


    class_value = car.find_elements_by_id('filter-class')
    for i in class_value:
        class_val = i.find_element_by_class_name('form-grid-box__details').text    

       
    

        
        info_dict = { 
         'career' : career_number,
         'class' :class_val,
         'last_10' : last_10_11,
         'winrate' : win_rate,
         'place' : place,
         'track' : track,
         'distance' : distance,
         'track/distance': track_distance,
         'good': good,
         'soft' : soft,
         'heavy': heavy,
         'wet' : wet,
         'first_up': firstup,
         'secondup': secondup


        }
    #print(med_info_dict)
        info.append(info_dict)
    

main_page_df = pd.DataFrame(info)
#cleaning the columns
main_page_df['field_condition'] = field_condition * len(main_page_df)
main_page_df.insert(0,'horse_name', name_2)
main_page_df['horse_name'] = main_page_df.horse_name.str[3:]

main_page_df.horse_name= main_page_df.horse_name.str.lower()
main_page_df['horse_name'] = main_page_df['horse_name'] .str.replace(r'^[.\s]*', '')
main_page_df['horse_name'] = main_page_df['horse_name'].str.strip()


driver.quit()
```

```python
driver = webdriver.Chrome(path)
driver.get(f'{website_ladbrokes}')
time.sleep(5)
gg = driver.find_elements_by_class_name('infohub-nav__list')
for ii in gg:
    element = ii.find_element_by_xpath("//*[contains(text(), 'Bet Tracker')]")
    element.click()
time.sleep(1)
horse_name = []
pound_sign = []
dollar_sign = []

market_container = driver.find_elements_by_class_name('race-table__body')
for i in market_container:
    main_runner = i.find_elements_by_css_selector("[class='race-runner-details']")
    for i in main_runner:
        horse = i.find_elements_by_class_name('runner-entrant-name')
        for x in horse:
            runner = x.find_element_by_class_name('runner-name').text
            horse_name.append(runner)

pound = driver.find_elements_by_class_name('number.show-icon.progress-track')
for i in pound:
    y = i.find_elements_by_class_name('progress-value')
    time.sleep(2)
    for i in y:
      pound_sign.append(i.get_attribute('style'))
#print(pound_sign)

dollar = driver.find_elements_by_class_name('dollar.show-icon.progress-track')
for i in dollar:
    y = i.find_elements_by_class_name('progress-value')
    time.sleep(2)
    for i in y:
      dollar_sign.append(i.get_attribute('style'))
#print(dollar_sign)



my_dict1 = {'horse_name': horse_name,
           'pound_sign': pound_sign,
           'dollar_sign' : dollar_sign}
#print(my_dict1)
time.sleep(1)
number_bets_df = pd.DataFrame(my_dict1)
number_bets_df.horse_name= number_bets_df.horse_name.str.lower()
number_bets_df.horse_name = number_bets_df.horse_name.str.strip()
driver.quit()
```

```python
# merging columns for a full dataframe
final_df = pd.merge(main_page_df, number_bets_df, on='horse_name', how='inner')
```
